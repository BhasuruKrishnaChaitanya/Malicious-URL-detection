```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('security.csv')
df.head()
df['TIPO'] = df['TIPO'].map({'Maligna': 1, 'Benigna': 0})
```

```python
pd.options.display.max_columns = 24
df.describe(include='all')
```

```python
df.drop('URL', axis=1, inplace=True)
df.drop('NUMBER_SPECIAL_CHARACTERS', axis=1, inplace=True)
df.drop('CHARSET', axis=1, inplace=True)
df.drop('SERVER', axis=1, inplace=True)
df.drop('CACHE_CONTROL', axis=1, inplace=True)
df.drop('CONTENT_LENGTH', axis=1, inplace=True)
df.drop('TCP_CONVERSATION_EXCHANGE', axis=1, inplace=True)
df.drop('DIST_REMOTE_TCP_PORT', axis=1, inplace=True)
df.drop('REMOTE_IPS', axis=1, inplace=True)
df.drop('APP_BYTES', axis=1, inplace=True)
df.drop('UDP_PACKETS', axis=1, inplace=True)
df.drop('SOURCE_APP_PACKETS', axis=1, inplace=True)
df.drop('REMOTE_APP_PACKETS', axis=1, inplace=True)
df.drop('SOURCE_APP_BYTES', axis=1, inplace=True)
df.drop('REMOTE_APP_BYTES', axis=1, inplace=True)
df.drop('APP_PACKETS', axis=1, inplace=True)
df.drop('DNS_QUERY_TIMES', axis=1, inplace=True)
```

```python
print(df.isnull().sum())
df[pd.isnull(df).any(axis=1)]
```

```python
df = df.interpolate()
print(df.isnull().sum())
```

```python
df.describe(include='all')
```

```python
dataset_with_dummies = pd.get_dummies(df,prefix_sep='--')
#print(dataset_with_dummies.head())
dataset_with_dummies.to_csv('d1.csv')
```
